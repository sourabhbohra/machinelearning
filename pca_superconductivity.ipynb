```python
from scipy.stats import zscore
from sklearn.preprocessing import Imputer
from sklearn.metrics import accuracy_score
import seaborn as sns
from scipy import stats
from sklearn import metrics
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
```

```python
data.shape
```

```python
data = data.dropna()
```

```python
data.shape
```

```python
data.head()
```

```python
response = data.iloc[:,81]
```

```python
preds = data.iloc[:,:81]
```

```python
preds.shape
```

```python
response.shape
```

```python
preds_z = preds.apply(zscore)
preds_z
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=81)
principalComponents = pca.fit_transform(X = preds_z)
```

```python
#data_comps = pd.DataFrame(data = principalComponents
                         ,columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13'])
```

```python
print(pca.explained_variance_ratio_.cumsum())
```

```python
len(pca.explained_variance_ratio_.cumsum())
```

```python
data_comps = pd.DataFrame(data = principalComponents)
```

```python
data_comps
```

```python
data_req = data_comps.iloc[:,:12]
```

```python
data_req
```
